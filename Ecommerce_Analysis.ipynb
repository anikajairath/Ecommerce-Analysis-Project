```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\admin\Desktop\Ecommerce Project\ecommerce.csv")
df.head()
```

```python
#Inspect the structure of the dataset

df.shape #rows and columns in the dataset
df.columns #column names of dataframes
df.info() #summary of dataset structure
df.describe() #statistical summary for numerical columns - mean, count, avg, etc
```

```python
df.isnull().sum() #checking for any null values
```

```python
df.duplicated().sum() #check for any duplicate rows
df.drop_duplicates(inplace=True) #drop duplicate rows if any
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date']) #Change the type of order_date to date instead of object
df.info()
```

```python
df['Revenue'] = df['Quantity'] * df['Price'] * (1 - df['Discount']/100) #Quantity x Price x Percentage of the price paid
df.head()
```

```python
df['Month'] = df['Order_Date'].dt.month #Create a new column Month by extracting it
```

```python
df['Revenue'].sum() #Get Total Revenue
```

```python
df.groupby('Product_Name')['Revenue'].sum().sort_values(ascending=False).head(10) #Top 10 products by revenue
```

```python
df.groupby('Region')['Revenue'].sum() #Region wide sales
```

```python
df.groupby('Month')['Revenue'].sum() #Monthly sales trend
```

```python
df['Customer_ID'].value_counts() #Repeat customers
```

```python
df.groupby('Payment_Mode')['Revenue'].sum() #Payment Method analysis
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Order_Date'] = df['Order_Date'].dt.strftime('%Y-%m-%d') #Changing the date type to yyyy-mm-dd for proper SQL use
```

```python
df.to_csv("ecommerce_cleaned.csv", index=False)
```
